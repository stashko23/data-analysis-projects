# Описание проекта
Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

**Цели проекта:** 

- Выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт.
- Спланировать рекламные кампании.

В ходе исследования **необходимо проверить гипотезы:**

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

## Описание данных

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

## Подготовка данных

### Переименование столбцов

Приведу все столбцы к нижнему регистру

```python
df.columns = df.columns.str.lower()
```

### Преобразование типов данных

```python
df.info()
```

Необходимо столбец `year_of_release` привести к целочисленному типу данных. Считаю, что к типу datetime нет смысла приводить, так как имеется только год релиза и в виде целого числа он имеет место быть в данном случае. Столбец `user_score` к float, так как оценки пользователей являются числами с плавающей запятой.

Начну со столбца `year_of_release`. Пропуски заполнять с помощью таких же игр, но с других платформ не имеет смысла, так как одна игра на разные платформы может выходить в разные годы. Количество пропусков в этом столбце небольшое, поэтому удалю их.

```python
# удаление пропусков
df.dropna(subset=['year_of_release'], inplace=True)

# приведу к целочисленному типу данных
df['year_of_release'] = df['year_of_release'].astype('int')

# посмотрю, какие остались значения в этом столбце
df['year_of_release'].sort_values().unique()
```

Теперь столбец `user_score`

```python
df['user_score'].sort_values().unique()
```

Для преобразования в тип данных float мешает значение 'tbd'. TBD - аббревиатура, означающая в данном контексте то, что оценка еще не определена. Думаю, что такие значения стоит изменить на NaN. 

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan

# теперь изменю тип данных на float
df['user_score'] = df['user_score'].astype('float')
```

### Поиск дубликатов

```python
# проверка на наличие явных дубликатов
df.duplicated().sum()
```

```python
# поиск неявных дубликатов по столбцам `name`, `platform`, `year_of_release`
df[df.duplicated(subset=['name', 'platform', 'year_of_release'], keep=False)]
```

```python
# нашлось 2 повтора, оставлю только первый из них
df = df.drop_duplicates(subset=['name', 'platform', 'year_of_release'], keep='first')
```

### Обработка пропусков

Сделаю анализ пропусков

```python
df.isnull().sum().sort_values(ascending=False) / df.shape[0] * 100
```

Что касается столбцов с оценками:` user_score` и `critic_score`. Заполнять такие столбцы средними/медианными и другими оценками в данном случае не стоит. Поэтому осталю пропуски как есть. Они могли возникнуть по причине того, что не все игры  популярны, какие-то из них созданы для узкого круга игроков, поэтому не все игры подвергаются оценкам. 

Посмотрю на значения из столбца `rating`

```python
df.rating.unique()
```

**Система ESRB**
- **EC** Для детей младшего возраста. Подходит для детей от 3 лет. 
- **E**	Подходит для всех возрастных категорий.  До 1998 года этот рейтинг назывался **К-А**
- **E10+**	Подходит для лиц старше 10 лет.
- **T**	Подходит для лиц старше 13 лет. 
- **M**	Подходит для лиц старше 17 лет.
- **AO** Подходит для взрослых старше 18 лет.
- **RP** Играм с категорией RP еще не присвоен окончательный рейтинг ESRB.

Пропусков много, лишние значения вне рейтинговой системы отсутствуют. Оценка `К-А` теперь имеет название `E`, поэтому изменю `К-А` на `E`.

```python
df.loc[df['rating'] == 'K-A', 'rating'] = 'E'
```

Пропусков в столбце `rating` порядка 40%. ESRB - это организация, которая принимает и определяет рейтинг для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде. Поэтому пропуски имеют признак того, что игра выпущена не в Северо-Американском регионе.

```python
df['rating'] = df['rating'].fillna('NR')
```

Пропуски в столбцах `genre` и `name` удалю.

```python
df.dropna(subset=['genre', 'name'], inplace=True)
```

### Подсчет суммарных продаж

Создам столбец `total_sales` для подсчета суммарных продаж

```python
df['total_sales'] = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis = 1)
```

### Анализ платформ

```python
# посмотрю на распределение значений суммарных продаж в зависимости от платформы
top_platforms = df.groupby('platform')['total_sales'].agg('sum').sort_values(ascending=False).to_frame()
top_platforms.describe()
```

Наибольшие суммарные продажи - те, что больше 300 млн., по ним найду топ-платформ

```python
top_platforms = top_platforms.query('total_sales > 300')

# строю график продаж у топ-8 платформ
top_platforms_sales = df.pivot_table(
                                        index = 'year_of_release', 
                                        columns = 'platform',
                                        values = 'total_sales', 
                                        aggfunc = 'sum'
                                     ) \
                        .fillna(0) 

top_platforms_sales = top_platforms_sales[top_platforms.index]
top_platforms_sales.plot(xticks=top_platforms_sales.index, figsize=(14,7))
plt.title('Распределение продаж у ТОП-8 платформ', loc='center') # название графика
plt.xlabel('Год релиза') # подпись оси x
plt.ylabel('Количество проданных игр, млн') # подпись оси y
plt.grid(linestyle='--', alpha=0.3)
plt.xticks(rotation = 45)
plt.show()
```

```python
# посчитаю среднюю продолжительность актуальности платформы
validity_top_platform = []
for name_platform in top_platforms_sales.columns:
    validity_top_platform.append(len(top_platforms_sales[top_platforms_sales[name_platform] != 0]))
print(f'Средняя продолжительность актуальности платформы {np.mean(validity_top_platform)} лет')
```

В среднем цикл появления и исчезновения платформы составляет примерно 10 лет. Из графика видно, что продажи начинают падать в среднем через 5 лет после релиза платформы. Поэтому, для прогноза продаж на 2017 год, осталю данные за последние 5 лет, т.к. 5 лет - это и не слишком большой, и не слишком маленький срок для построения прогноза.

```python
df = df.query('year_of_release > 2011')
```

### Влияние отзывов пользователей и критиков на продажи внутри популярных платформ 

Для каждой потенциально прибыльной платформы посчитаю среднее значение `critic_score` и `user_score`

```python
# отоберу 5 лидирующих платформ
leader_platform = df.query("platform in ['PS4', 'PS3', 'X360', '3DS', 'XOne']")
leader_platform.pivot_table(index='platform', values=('user_score', 'critic_score'))
```

Оценки имеют небольшой разброс в значениях и у `critic_score` и у `user_score`. Средние значения не маленькие, но и не слишком высокие, немногим больше среднего по шкале оценки. Сделаю диаграммы рассеяния и посчитаю коэффициенты корелляции для каждой из платформ.

```python
def create_scatter_corr(name_platform):
    score_platform = leader_platform.query('platform == @name_platform') 
    fig, ax = plt.subplots(1, 2, figsize=(10,4))
    score_platform.plot(x='total_sales', y='user_score', kind='scatter', alpha=0.5, c = 'b', s = 4, ax=ax[0])  
    score_platform.plot(x='total_sales', y='critic_score', kind='scatter', alpha=0.5, c = 'r', s = 4, ax=ax[1])
    fig.suptitle(f'Зависимость продаж платформы {name_platform}\n от оценок пользователей и критиков')
    ax[0].set(xlabel='Количество проданных игр, млн', ylabel='Оценки пользователей')
    ax[1].set(xlabel='Количество проданных игр, млн', ylabel='Оценки критиков')
    plt.show()

    plt.figure(figsize=(5,1))
    sns.heatmap(
                    score_platform[['user_score', 'critic_score', 'total_sales']] \
                    .corr()[['total_sales']] \
                    .sort_values(by='total_sales', ascending=False), 
                     cmap="Reds", annot=True, annot_kws={'size':10}
                )
    
    plt.title(f'\nКорреляция отзывов и продаж\n для платформы {name_platform}', loc='center', fontsize=12)
    
    
for name_plt in leader_platform['platform'].unique():
    create_scatter_corr(name_plt)
```

Что касается оценок критиков и продаж - значения варьируются от 0.32 до 0.42. Оценки критиков имеют бОльшее влияние на продажи, относительно оценок пользователей. ## Портрет пользователя каждого региона

```python
def user_portret(sign, region, x):
    region_dict = {'na_sales': 'Продажи в Северной Америке', 'eu_sales': 'Продажи в Европе', 'jp_sales': 'Продажи в Японии'}
    sign_dict = {'platform': 'Платформы', 'genre': 'Жанры', 'rating': 'Рейтинг ESRB'}
    top_5 = df.groupby(sign)\
              .agg({region: 'sum'}) \
              .sort_values(by=region, ascending=False) \
              .apply(lambda x: x / x.sum())
    top_5.loc[len(top_5.index)] = sum(top_5[region][5:])
    
    if top_5.loc[len(top_5.index)-1][region] != 0:
        top_5 = top_5.rename(index={len(top_5.index)-1: 'Other'})
        index_5 = top_5[:5].index.to_list()
        index_5.append('Other')
        top_5 = top_5.query('index in @index_5')
    else:
        top_5.drop(index=len(top_5.index)-1, inplace=True)

    top_5.plot(kind='pie', ax=ax[x], subplots=True, fontsize=12, autopct='%.0f%%', \
              legend=False, cmap='Set3')
    ax[0].set(title='Северная Америка', ylabel='')
    ax[1].set(title='Европа', ylabel='')
    ax[2].set(title='Япония', ylabel='')
    fig.suptitle(f'{sign_dict[sign]} в разных регионах', fontsize=20)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18,6))

user_portret('platform', 'na_sales', 0)
user_portret('platform', 'eu_sales', 1)
user_portret('platform', 'jp_sales', 2)
```

В Америке самыми популярными платформами являются X360, 40% от всех платформ занимают платформы от Xbox. В Европе самыми популярными являются платформы PS4, почти 50% от всех платформ занимают платформы от Sony. В Японии 46% платформ - это Nintendo 3DS

## Проверка гипотез

Для проверки гипотез буду использовать **t-тест** (если обе генеральные совокупности не очень большие, можно рассчитать их средние и сделать вывод, какое больше, но, вывод о значимости этой разницы всё равно не сделать без статистического теста.)

В обоих случаях буду проверять гипотезы о равенстве средних двух генеральных совокупностей между собой (равны или не равны).

Для этого В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: `scipy.stats.ttest_ind`

Для этого нужно учесть требования к t-тесту: 
- Генеральные совокупности не должны зависеть друг от друга. (выборки для обоих гипотез независимы между собой, поэтому требование выполняется)
- Выборочные средние должны быть нормально распределены.
- Дисперсии рассматриваемых генеральных совокупностей должны быть равны (если это не так, то необходимо параметру `equal_var` присвоить значение False.)

Создам переменные `xone` и `pc`, в которых будут храниться оценки пользователей для платформ XOne и PC

```python
xone = df.query('platform == "XOne"')['user_score'].dropna().reset_index(drop=True)
pc = df.query('platform == "PC"')['user_score'].dropna().reset_index(drop=True)
```

Создам переменные `action` и `sports`, в которых будут храниться оценки пользователей для жанров Action и Sports

```python
action = df.query('genre == "Action"')['user_score'].dropna().reset_index(drop=True)
sports = df.query('genre == "Sports"')['user_score'].dropna().reset_index(drop=True)
```

### «Средние пользовательские рейтинги платформ Xbox One и PC одинаковые»

Посмотрю как распределены выборочные средние для Xbox One и PC 

```python
def normal_distribution(data_list):
    for name_platform in data_list:
        
        name_dict = {
                        len(xone): 'платформы XOne', len(pc): 'платформы PC', \
                        len(action): 'жанра Action', len(sports): 'жанра Sports'
                     }
        name = name_dict[len(name_platform)]
        # здесь будут все возможные выборки из ГС длиной меньше на 1, чем длина ГС 
        all_combinations = []
        for comb in itertools.combinations(name_platform, len(name_platform) - 1):
            all_combinations.append(comb)
            
        # здесь будут средние значения каждой выборки
        comb_means = []
        for comb in all_combinations:
            # считаем среднее для каждой получившейся выборки и добавляем в список
            comb_means.append(sum(comb) / len(comb))  
            
        comb_means = pd.Series(comb_means)
    
        plt.figure(figsize=(5, 3))
        comb_means.hist(bins=10, grid=False)
        plt.title(f'График распределения выборочных средних\n для {name}', loc='center') # название графика
        plt.xlabel(f'Распределения выборочных средних оценок\n пользователей для {name}') 
        plt.ylabel('Количество оценок')
        plt.show()

        print(f'Средняя оценка пользователей для {name}: {name_platform.mean().round(2)}')
        print(f'Количество оценок пользователей для {name} {len(name_platform)}\n')
    
    
normal_distribution([xone, pc])
```

Значения распределены нормально, но количество оценок для Xbox One и PC отличается, так же есть сомнения в том, что дисперсии равны, поэтому параметру `equal_var` присвою значение False. 

**Сформулирую нулевую гипотезу:**
- **H₀** = Средние пользовательские рейтинги платформ Xbox One и PC равны

**Исходя из нулевой гипотезы сформулирую альтернативную:**
- **H₁** = Средние пользовательские рейтинги платформ Xbox One и PC не равны

Уровень статистической значимости **α** = 0.01

```python
# уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = 0.01

results = st.ttest_ind(xone, pc, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Значение p-value показывает, что вероятность получить такой или еще более далекий от предполагаемого результат равна ~55%. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними оценками пользователей для платформ Xbox One и PC. Нулевую гипотезу точно не нужно отвергать.

**Гипотеза «Средние пользовательские рейтинги платформ Xbox One и PC одинаковые» подтвердилась**

### «Средние пользовательские рейтинги жанров Action и Sports разные»

```python
normal_distribution([action, sports])
```

Значения распределены нормально, но количество оценок для жанров Action и Sports отличается, так же есть сомнения в том, что дисперсии равны, поэтому параметру `equal_var` присвою значение False. 

**Сформулирую нулевую гипотезу:**
- **H₀** = Средние пользовательские рейтинги жанров Action и Sports равны

**Исходя из нулевой гипотезы сформулирую альтернативную:**
- **H₁** = Средние пользовательские рейтинги жанров Action и Sports не равны

Уровень статистической значимости **α** = 0.01

```python
# уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = 0.01

results = st.ttest_ind(action, sports, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Значение p-value показывает, что вероятность получить такой или еще более далекий от предполагаемого результат равна 5.99e-20. Это крайне маленькая вероятность, чтобы делать вывод о равенстве средних оценок пользователей для жанров Action и Sports. Нулевую гипотезу отвергаю, так как есть значимые различия между средними оценками пользователей выбранных жанров.

**Гипотеза «Средние пользовательские рейтинги жанров Action и Sports разные» подтвердилась**
